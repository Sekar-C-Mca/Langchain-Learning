```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_community.embeddings import OllamaEmbeddings

##Load Data;
loader=TextLoader("DataIngestion/speech.txt")
document=loader.load()
document
chunk = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=30)
chunk_text=chunk.split_documents(document)
chunk_text

embedding = OllamaEmbeddings(model="nomic-embed-text")

vector_db = Chroma.from_documents(
    documents=chunk_text,
    embedding=embedding,
    persist_directory="./chroma_db"
)
```

```python
vector_db
```

```python
vector=vector_db.similarity_search("easier for us to conduct ourselves")
```
